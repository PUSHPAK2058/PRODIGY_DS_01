```python
import pandas as pd

# Sample data for illustration
data = {
    'Country Name': ['India', 'China', 'USA', 'Indonesia', 'Pakistan'],
    '2022': [1400, 1440, 331, 276, 243]
}
df_cleaned = pd.DataFrame(data)

# Verify structure
print(df_cleaned.head())

# Extract and rename columns
df_2022 = df_cleaned[['Country Name', '2022']].dropna()
df_2022.columns = ['country', 'population']

# Preview result
print(df_2022.head())
```

```python
# Define df_cleaned first (example structure)
df_cleaned = pd.DataFrame({
    'Country Name': ['India', 'China', 'USA', 'Indonesia', 'Pakistan'],
    '2022': [1400, 1440, 331, 276, 243]
})

# Extract and rename columns
df_2022 = df_cleaned[['Country Name', '2022']].dropna()
df_2022.columns = ['Country', 'Population']

# Plot the data
import matplotlib.pyplot as plt
top_10 = df_2022.sort_values(by='Population', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_10['Country'], top_10['Population'], color='skyblue')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xticks(rotation=45)
plt.ylabel('Population')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df_2022['Population'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Population Across Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
